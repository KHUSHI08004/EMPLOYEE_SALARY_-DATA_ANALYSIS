```python
import pandas as pd

df = pd.read_csv("e_data.csv")

print(df.head())
print(df.shape)
```

```python
print(df.columns)
```

```python
# Average Salary
print("Average Salary:", df['Monthly_Salary'].mean())

# Highest Salary
print("Highest Salary:", df['Monthly_Salary'].max())

print("Highest Paid Employee:")
print(df.loc[df['Monthly_Salary'].idxmax()])

# Department-wise Salary
print("\nDepartment-wise Salary:")
print(df.groupby('Department')['Monthly_Salary'].mean())

# Experience vs Salary
print("\nCorrelation:")
print(df[['Experience_Years','Monthly_Salary']].corr())
```

```python
import matplotlib.pyplot as plt

# Histogram
df['Monthly_Salary'].plot(kind='hist')
plt.title("Salary Distribution")
plt.show()

# Bar Chart
df.groupby('Department')['Monthly_Salary'].mean().plot(kind='bar')
plt.title("Department-wise Salary")
plt.show()

# Scatter Plot
plt.scatter(df['Experience_Years'], df['Monthly_Salary'])
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("Experience vs Salary")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate total salary by department
dept_salary = df.groupby('Department')['Monthly_Salary'].sum()

# Pie chart
dept_salary.plot(kind='pie', autopct='%1.1f%%')
plt.title("Department-wise Salary Distribution")
plt.ylabel("")   # removes extra label
plt.show()
```
